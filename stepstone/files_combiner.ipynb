get 2 excel files 

get one of excel files and start from bottom to up sequentially to check if certain feature has '___' value and remove that row from dataset and exit if it doesn't equal to '___'

```python
import pandas as pd
import os

# Define a function to read and preprocess each file
def read_and_process_file(file_name):
    file_path = os.path.join('data', file_name)
    df = pd.read_excel(file_path)
    # Rename the feature 'Language' to 'Profession'
    df.rename(columns={'Language': 'Profession'}, inplace=True)
    # Set the value of 'Profession' to the first two letters in the name of the file
    profession_value = file_name[:2]
    df['Profession'] = profession_value
    return df

# Read and preprocess each file
file_names = ['DE_scraped_1-175.xlsx', 'DE_scraped_175-351.xlsx', 
              'DE_scraped_351-522.xlsx', 'DE_scraped_522-700.xlsx', 'DE_scraped_700-all.xlsx',
              'DS_scraped_1-180.xlsx', 'DS_scraped_180-all.xlsx', 'scraped_again.xlsx', 'scraped_again_2.xlsx']

dataframes = [read_and_process_file(file_name) for file_name in file_names]

# Combine the preprocessed dataframes
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df_backup = combined_df.copy()

placeholder = '________________________'

# Drop all rows where 'Job' feature equals '___'
combined_df = combined_df[combined_df['Job'] != placeholder]
combined_df_backup = combined_df_backup[combined_df_backup['Job'] == placeholder]

# Drop duplicates in combined_df based on 'Link' feature
combined_df = combined_df.drop_duplicates(subset='Link')

# Save modified dataframe to excel
combined_df.to_excel('dataset_stepstone.xlsx', index=False)

# Remove duplicates in combined_df_backup based on 'Link' feature
combined_df_backup = combined_df_backup.drop_duplicates(subset='Link')

# Remove rows in combined_df_backup if Link already exists in combined_df
unscrapable_links = combined_df_backup[~combined_df_backup['Link'].isin(combined_df['Link'])]

# Save the modified backup dataframe to excel
unscrapable_links.to_excel('unscrapable_links.xlsx', index=False)
```
